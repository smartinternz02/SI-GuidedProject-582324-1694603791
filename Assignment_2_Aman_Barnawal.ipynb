Importing dependencies

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('car_crashes')
```

```python
data.head()
```

SCATTERPLOT

```python
sns.scatterplot(data,x="alcohol",y="total")
plt.title("Alcohol vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
plt.show()
```

Interference:

*   Scatterplot shows relationship between alcohol consumption and total car crashes
*   There is a positive correlation between consumption and crashes as total number of crashes increase with increase in alcohol consumption



PAIRPLOT

```python
sns.pairplot(data)
plt.title("Pairplot of Numeric Variables")
plt.show()
```

Inference:

*   Provides relationship and patterns between all numeric variables in the dataset



BARPLOT

```python
plt.figure(figsize=(15,6))
sns.barplot(data,x="abbrev",y="total")
plt.title("Number of Crashes by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Total Crashes")

plt.xticks(rotation=90)
plt.show()
```

Inference:

*   Displays total crasehs in each US state
*   DC has lowest while ND, SC has highest number of crashes






BOXPLOT

```python
plt.figure(figsize=(16,6))
sns.boxplot(data, x="speeding", y="total")
plt.title("Box Plot of Number of Crashes by Speeding")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")


plt.xticks(rotation=90)
plt.show()
```

Inference:

* The box plot illustrates the distribution of the total number of crashes in states with and without significant speeding as a factor.
* States with significant speeding tend to have a wider range of total crashes, including some outliers with exceptionally high crash counts.





HISTPLOT

```python
sns.histplot(data, x="alcohol", kde=True)
plt.title("Histogram of Alcohol Consumption")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Frequency")
plt.show()
```

Inference:

*   This displays the distribution of alcohol consumption percentages among states
*   It indicates that a majority of states have lower alcohol consumption percentages, with a peak around 10-12%.





REGRESSION PLOT

```python
sns.regplot(data, x="speeding", y="alcohol")
plt.title("Speeding vs. Alcohol Consumption")
plt.xlabel("Speeding")
plt.ylabel("Alcohol Consumption")
plt.show()
```

Inference:


*   The regression plot suggests a positive relationship between speeding and alcohol consumption during car crashes



HEATMAP

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Inference:

*    Helps identify strong and weak correlations between different factors related to car crashes
